```python
import math
# Constants
C = 50
H = 30

# Read input
D_values = input().split(',')

# Calculate Q for each D
result = []
for D in D_values:
    Q = math.sqrt((2 * C * int(D)) / H)
    result.append(str(round(Q)))

# Print output
print(",".join(result))
```

```python
def check_list(nums):
    return nums.count(19) == 2 and nums.count(5) >= 3

# Input list
numbers = list(map(int, input().split()))

# Output
print(check_list(numbers))
```

```python
def find_target_indices(matrix, target):
    result = []
    
    for i, row in enumerate(matrix):
        for j, value in enumerate(row):
            if value == target:
                result.append([i, j])
    
    return result


# Input
matrix = [
    [1, 3, 2, 32, 19],
    [19, 2, 48, 19],
    [],
    [9, 35, 4],
    [3, 19]
]

target = 19

# Output
print(find_target_indices(matrix, target))
```
